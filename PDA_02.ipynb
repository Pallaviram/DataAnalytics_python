```python
import matplotlib.pyplot as plt

sub = ["ADA", "AJ", "DBMS"]
Pri = [77, 66, 78]
Ammu = [78, 88, 66]
plt.subplot(1, 2, 1)
plt.bar(sub, Pri, color="k", label="Pri scores",width=1,align="center")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Pri's Scores")
plt.legend()

plt.subplot(1, 2, 2)
plt.bar(sub, Ammu, color="b", label="Ammu scores",width=0.5,align="edge")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Ammu's Scores")
plt.legend()

plt.show()


```
